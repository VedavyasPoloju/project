```python
X = data.loc[:,'age':'histology']
y = data.loc[:,['class']]
```

```python
from sklearn import preprocessing
lb = preprocessing.LabelBinarizer()
y = lb.fit_transform(y)
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
X = SS.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()

model_rf.fit(X_train, y_train)        
model_rf.score(X_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
# first neural network with keras tutorial
from sklearn.metrics import accuracy_score, confusion_matrix,precision_score, recall_score, roc_auc_score,roc_curve, auc, f1_score 
predictions = model_rf.predict(X_test)

acc_RF = accuracy_score(y_test, predictions)*100
print('Accuracy of Random Forest (%): \n',acc_RF)
      

pre_RF = precision_score(y_test,predictions)
print('Precision Score (%): \n',pre_RF )
rec_RF = recall_score(y_test,predictions)
print('Recall Score (%): \n', rec_RF)
```

```python
clf = MLPClassifier(solver='lbfgs',hidden_layer_sizes=10,batch_size=200,max_iter=200, random_state=1)
clf.fit(X_train,y_train)
predictions = clf.predict(X_test)

acc_NN = accuracy_score(y_test, predictions)*100
print('Accuracy of MLP/Neural Network (%): \n',acc_NN)
      
pre_NN = precision_score(y_test,predictions)
print('Precision Score (%): \n',pre_NN )
rec_NN = recall_score(y_test,predictions)
print('Recall Score (%): \n', rec_NN)
```

```python
from sklearn.linear_model import Perceptron
pp = Perceptron()
modelo = pp.fit(X_train, y_train)
y_predict = modelo.predict(X_test)
```

```python
score = modelo.score(X_test, y_test)
score
```

```python
predictions = modelo.predict(X_test)

acc_NN = accuracy_score(y_test, predictions)*100
print('Accuracy of MLP/Neural Network (%): \n',acc_NN)
      
pre_NN = precision_score(y_test,predictions)
print('Precision Score (%): \n',pre_NN )
rec_NN = recall_score(y_test,predictions)
print('Recall Score (%): \n', rec_NN)
```

```python

#KNN Classifier with 3 neighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
classifier_KNN = KNeighborsClassifier(n_neighbors=3)
classifier_KNN.fit(X_train, y_train)
print('Accuracy of KNN classifier on test set: {:.2f}'.format(classifier_KNN.score(X_test, y_test)))
y_pred=classifier_KNN.predict(X_test)
#Classification Report
print(classification_report(y_test, y_pred, target_names=["0","1"]))
```

```python
from sklearn import svm

svc = svm.SVC(kernel='linear', C=1).fit(X_train, y_train)

y_pred = svc.predict(X_test)
from sklearn.metrics import confusion_matrix
confusion_matrix1 = confusion_matrix(y_test, y_pred)
print(confusion_matrix1)
print(classification_report(y_test, y_pred, target_names=["0","1"]))
```
